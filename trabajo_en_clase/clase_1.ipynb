# Hacer un primer grafico con matplotlib y seaborn

```python
import pandas
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x:list[str] = ['EDA','Data Wrangling','Mineria datos','Limpieza Datos','SQL','Est. Inferencial','ML','DL','IA']
y:list[int] = [5,10,15,20,30,25,31,30,20]
```

```python
data_dict:dict = {
    "Caracteristicas": x,
    "Frecuencia": y
}

data_dict
```

```python
df = pandas.DataFrame(data_dict) #dataFrame
```

```python
df
```

```python
plt.figure(figsize=(16,9))
sns.barplot(df, x="Caracteristicas", y="Frecuencia")
plt.show()
```
