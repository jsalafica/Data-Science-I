```python
# Generar data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,200)
y = np.tanh(x)

data = pd.DataFrame({'x':x,'y':y})
data.head()
```

```python
# que y donde voy a colocar esos datos
plt.figure(figsize=(8,4))

# los datos y que voy colocar
plt.plot(data["x"],data["y"], label="y = tanh(x)", color=(0.1, 0.2, 0.5, 0.3),linestyle=":")
plt.title("Grafico de funcion Tangencial")

# legendas, informacion adicional al grafico
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")

plt.text(5,0.5,r"$\sigma(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}}$")

plt.grid(True, alpha=0.5)
plt.tight_layout(pad=2)
plt.legend()
plt.show()
```

Cuadratica

```python
# Generar datos de entrada
x = np.linspace(-10, 10, 200)  # 200 valores entre -10 y 10

# Definir una relaci칩n cuadr치tica: y = ax^2 + bx + c
a, b, c = 2, -3, 5
y = a * x**2 + b * x + c

cuadradica = pd.DataFrame({'x':x,'y':y})
cuadradica.head()


# Visualizar los datos
plt.figure(figsize=(8, 5))

plt.plot(x, y, label=f"${a}x^2 + {b}x + {c}$")
plt.title("Curva Cuadr치tica")
plt.xlabel("x")
plt.ylabel("f(y)")
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.legend()
plt.grid(True)
plt.show()
```

```python
np.random.seed(42)
x = np.random.rand(50)
y = 2 * x + 1 + np.random.normal(scale=0.1, size=50)

data = pd.DataFrame({'x':x,'y':y})
data.head()
```

```python
fig, axes =plt.subplots(1, 3, figsize=(12,4))

axes[0].scatter(data["x"],data["y"], color="darkorange")
axes[0].grid(True)


axes[1].hist(data["y"], bins=10, color="darkorange", alpha=0.7)
axes[1].grid(True)

axes[2].plot(cuadradica.x, cuadradica.y, label=f"${a}x^2 + {b}x + {c}$")
axes[2].set_title("Curva Cuadr치tica")
axes[2].set_xlabel("x")
axes[2].set_ylabel("f(y)")

plt.tight_layout(pad=2)
plt.show()
```
